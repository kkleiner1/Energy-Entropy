```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv("energies.csv")
df['basis'] = pd.Categorical(df['basis'],['ccpvdz','ccpvtz','ccpvqz','ccpv5z'])

df_vmc = df[(df.method=='VMC')]
df_vmc_true = df_vmc[df_vmc.orbital_optimize==True]
df_vmc_false = df_vmc[df_vmc.orbital_optimize==False]
df_vmc_false.sort_values(by=['bond length'])
df_opt = df[df.orbital_optimize==True]
```

```python
g = sns.FacetGrid(hue='orbital_optimize', col='basis', data =df_vmc.sort_values(by=['bond length']))
g.map(plt.errorbar, 'bond length', 'energy', 'error', marker='o')
g.add_legend()
```

```python
g = sns.FacetGrid(row='method', hue='basis', data =df, sharey=False)
g.map(plt.errorbar, 'bond length', 'energy', 'error', marker='o')
g.add_legend()
```

```python
h = sns.FacetGrid(col='basis', hue='method', data =df_opt.sort_values(by=['bond length']))
h.map(plt.errorbar, 'bond length', 'energy', 'error', marker='o')
h.add_legend()
```
